```python
import os
from neo4j import GraphDatabase
from dotenv import load_dotenv


load_dotenv()
```

```python
NEO4J_USERNAME = os.getenv("NEO4J_USERNAME", "<instance-id>")
NEO4J_PASSWORD = os.getenv("NEO4J_PASSWORD", "<password>")
NEO4J_URI = os.getenv("NEO4J_URI", f"neo4j+s://{NEO4J_USERNAME}.databases.neo4j.io")
```

```python
driver = GraphDatabase.driver(NEO4J_URI, auth=(NEO4J_USERNAME, NEO4J_PASSWORD))
driver.verify_connectivity()
```

```python
records, summary, keys = driver.execute_query( # (1)
    "RETURN COUNT {()} AS count"
)

# Get the first record
first = records[0]      # (2)

# Print the count entry
print(first["count"])   # (3)
```

```python
cypher = """
MATCH (p:Person {name: $name})-[r:ACTED_IN]->(m:Movie)
RETURN m.title AS title, r.role AS role
"""
name = "Tom Hanks"

records, summary, keys = driver.execute_query( # (1)
    cypher,    # (2)
    name=name  # (3)
)
```

```python
movie = "Toy Story"

records, summary, keys = driver.execute_query("""
MATCH path = (person:Person)-[actedIn:ACTED_IN]->(movie:Movie {title: $title})
RETURN path, person, actedIn, movie
""", title=movie)
```
